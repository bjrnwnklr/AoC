# Data Cleaning and Preprocessing

From Pluralsight course Data Science Literacy - Representing, Processing and Preparing Data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('mobile_price_classification.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
data = data.drop('id', axis=1)

data.shape
```

```python
data.describe()
```

```python
data = data.rename(columns = {'blue': 'bluetooth', 
                             'fc': 'fc_megapixel',
                             'pc': 'pc_megapixel',
                             'm_dep': 'm_depth'})
```

```python
# identify duplicates
dupes = data.duplicated()

dupes.sum()
```

```python
# get rid of duplicates
data = data.drop_duplicates()

data.shape
```

```python
# find missing values
data.isna().sum()
```

```python
# fill with zero values
data['fc_megapixel'] = data['fc_megapixel'].fillna(0)

data.isna().sum()
```

```python
# find unique values
len(data['ram'].unique())
```

```python
# backfill (replace with next value in column)
data['ram'] = data['ram'].fillna(method='backfill')

len(data['ram'].unique())
```

```python
# fill with median value
data['mobile_wt'] = data['mobile_wt'].fillna(data['mobile_wt'].median())
```

```python
data.isna().sum()
```

```python
# drop any records with missing fields
data = data.dropna()

data.isna().sum()
```

```python
data.shape
```

```python
# write to file
data.to_csv('mobile_data_cleaned.csv', index=False)
```

## Identifying and removing outliers

```python
# drop numeric columns into a separate frame
numeric_data = data.drop(['bluetooth', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi', 'price_range'], axis=1)

numeric_data.head()
```

```python
# get categorical data columns into separate frame
categorical_data = data[['bluetooth', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi', 'price_range']]

categorical_data.head()
```

### Visualize outliers

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

#### boxplot

- middle line: mean
- box: interquartile range: 25 - 75 quartile
- whiskers: 1.5 times interquartile range from mean

```python
# show outliers in ram
fig, ax = plt.subplots(figsize=(10, 8))

sns.boxplot(numeric_data['ram'], orient='v')
```

```python
# visualize all in a single boxplot
fig, ax = plt.subplots(figsize=(10, 8))

bp = sns.boxplot(data=numeric_data)
bp.set_xticklabels(bp.get_xticklabels(), rotation=90)
```

### Standard scale to detect outliers

Use zscores to detect outliers (= how many standard deviations away from the mean)

- scale data to standard scale - to zero mean and unit variance

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_array = scaler.fit_transform(numeric_data)
```

```python
# create new dataframe with scaled data
scaled_data = pd.DataFrame(scaled_array, columns=numeric_data.columns)
scaled_data.head()
```

```python
# all mean values are very close to 0, all std are close to 1
scaled_data.describe()
```

```python
# data is now much easier to compare, outliers clearly visible
fig, ax = plt.subplots(figsize=(10, 8))

bp = sns.boxplot(data=scaled_data)
bp.set_xticklabels(bp.get_xticklabels(), rotation=90)
```
