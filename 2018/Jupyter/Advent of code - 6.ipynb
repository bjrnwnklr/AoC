# part 2

On the other hand, if the coordinates are safe, maybe the best you can do is try to find a region near as many coordinates as possible.

For example, suppose you want the sum of the Manhattan distance to all of the coordinates to be less than 32. For each location, add up the distances to all of the given coordinates; if the total of those distances is less than 32, that location is within the desired region.

Correct answer: 42998

```python
import collections as coll
#coordinates = [(1, 1), (1, 6), (8, 3), (3, 4), (5, 5), (8, 9)]
coordinate_file = open(r'D:\Python\Advent\6.1\input.txt', 'r').readlines()
coordinates_str = [line.split(',') for line in coordinate_file]
coordinates = [(int(x), int(y)) for x, y in coordinates_str]

print(coordinates)

dist_dict = coll.defaultdict(int)

x_min = min(c[0] for c in coordinates)
x_max = max(c[0] for c in coordinates)
y_min = min(c[1] for c in coordinates)
y_max = max(c[1] for c in coordinates)

print(x_max, y_max)
```

```python
threshold = 10000
within_region = set()

for x in range(x_min - 1, x_max + 1):
    for y in range(y_min - 1, y_max + 1):
        dist = [(abs(i - x) + abs(j - y)) for (i, j) in coordinates]
        if sum(dist) < threshold:
            within_region.add((x, y))

print(len(within_region))
print(within_region)
```
